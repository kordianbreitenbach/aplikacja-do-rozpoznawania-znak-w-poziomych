```python
import yaml

# Path to the file
file_path = '/content/drive/My Drive/Colab Notebooks/archive/data.yaml'

# Open and read the YAML file
with open(file_path, 'r') as file:
    data = yaml.safe_load(file)
```

```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
```

```python
model.train(data=file_path, seed=42, epochs=50)
```

```python
# Load the YOLO model (pre-trained weights can be used)
model = YOLO('yolov8n.pt')  # Replace 'yolov8n.pt' with other weights for larger models (e.g., 'yolov8m.pt')

# Train the model
model.train(data='/content/drive/My Drive/Colab Notebooks/archive/data.yaml',  # Path to data.yaml
            epochs=50,  # Number of epochs
            imgsz=640,  # Image size
            batch=16)   # Batch size
```

```python
metrics = model.val()  # This will validate the model using the validation dataset
print(metrics)
```

```python
new_class_names = ['Pas dla autobusów', 'Żółte oznaczenie', 'Linia 1', 'Linia 2',
                   'Przejście dla pieszych', 'Romboid', 'Zwolnij', 'Strzałka w lewo',
                   'Strzałka do przodu', 'Strzałka do przodu i w lewo',
                   'Strzałka do przodu i w prawo', 'Strzałka w prawo', 'Rower']
```

```python
results = model('/content/drive/My Drive/Colab Notebooks/archive/test/images/940d_jpg.rf.0ea83fb3761ecdb8b2933524d396e14a.jpg')  # Replace with the path to your test image
# Update class names in the results
for result in results:
    # Map old class IDs to new class names
    result.names = {i: name for i, name in enumerate(new_class_names)}
    print(result.names)  # Verify the updated names
result = results[0]
result.show()  # Show predictions
result.save()  # Save the predictions
```
